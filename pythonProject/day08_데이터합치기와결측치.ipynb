## 데이터 합치기

### 가로로 합치기

````python
import pandas as pd
````

````python
# 중간고사 데이터 만들기
test1 = pd.DataFrame({'id':[1, 2, 3, 4, 5],
                      'midterm':[60, 80, 70, 90, 85]})
test1
````

````python
test2 = pd.DataFrame({'id':[1, 2, 3, 4, 5],
                      'final':[70, 83, 65, 95, 80]})
test2
````

````python
# id 를 기준으로 합쳐서 total 에 할당
total = pd.merge(test1, test2, how = 'left', on = 'id')
total
````

### Q1. popadults는 해당 지역의 성인 인구, poptotal은 전체 인구일 때, midwest 데이터에 '전체 인구 대비 미성년 인구 백분율' 변수 추가

````python
# midwest 데이터 불러오기
midwest = pd.read_csv('./input/midwest.csv')

# 아래 처럼 하면 ratio 가 midwest에 저장되지 않은 상태라 2번문제 바로 풀기 힘듦
# midwest.assign(ratio = (midwest['poptotal'] - midwest['popadults']) / midwest['poptotal'] * 100)

# midwest에 백분율 변수 추가
midwest['ratio'] = (midwest['poptotal'] - midwest['popadults']) / midwest['poptotal'] * 100
midwest
````

### Q2. 미성년 인구 백분율이 가장 높은 상위 5개 county(지역)의 미성년 인구 백분율 출력

````python
# ratio 내림차순 정렬, 상위 5행 추출, 변수 추출
midwest.sort_values('ratio', ascending = False)\
    .head()\
    [['county', 'ratio']]
````

### Q3. 분류표의 기준에 따라 미성년 비율 등급 변수를 추가하고 각 등급에 몇 개의 지역이 있는지 확인

````python
import numpy as np
# midwest 에 grade 변수 추가
midwest['grade'] = np.where(midwest['ratio'] >= 40, 'large',\
                    np.where(midwest['ratio'] >= 30, 'middle', 'small'))

# 방법1
# grade 별 분리, grade 빈도 구하기
midwest.groupby('grade')\
    .agg(n = ('grade', 'count'))
````

````python
# 방법2
midwest['grade'].value_counts()
````

### Q4. popasian은 해당 지역의 아시아인 인구를 나타낼 때 '전체 인구 대비 아시아인 인구 백분율' 변수를 추가하고 하위 10개 지역의 state(주), county(지역), 아시아인 인구 백분율 출력

````python
midwest.assign(ratio_asian = midwest['popasian'] / midwest['poptotal'] * 100)\
    .sort_values('ratio_asian')\
    .head(10)\
    [['state', 'county', 'ratio_asian']]
````

# 결측치 정제하기

## 결측치 만들기

````python
df = pd.DataFrame({'sex':['M', 'F', np.nan, 'M', 'F'],
                   'score':[5, 4, 3, 4, np.nan]})
df
````

````python
df['score'] + 1 # 연산하면 출력 결과도 NaN
````

### 결측치 확인하기

````python
pd.isna(df)
````

````python
pd.isna(df).sum()   # 결측치 빈도(개수) 확인
````

### 결측치 있는 행 제거하기

````python
df_nomiss = df.dropna(subset = ['score'])
````

````python
df_nomiss['score'] + 1  # 결측치 없이 연산 가능
````

### 여러 변수에 결측치 없는 데이터 추출하기

````python
df_nomiss2 = df.dropna(subset = ['score', 'sex'])
df_nomiss2
````

### 결측치가 하나라도 있으면 제거하기

````python
df_nomiss3 = df.dropna()    # 모든 변수에 결측치 없는 데이터 추출
df_nomiss3
````

### 결측치 제거하지 않고 분석하기(함수가 자동으로 제거)

````python
df.groupby('sex').agg(mean_score = ('score', 'mean'),
                      sum_score = ('score', 'sum'))
````

### Q0. mpg 데이터를 불러와 일부러 몇 개의 값을 결측치로 만들기

### Q1. drv(구동 방식)별로 hwy(고속도로 연비) 평균이 어떻게 다른지 확인하기 위해 drv 변수와 hwy 변수에 결측치가 몇개 있는지 확인

````python
# drv 결측치 빈도 확인
mpg_new['drv'].isna().sum()
````

````python
# hwy 결측치 빈도 확인
mpg_new['hwy'].isna().sum()
````

````python
# drv, hwy 결측치 빈도 일괄 확인
mpg_new[['drv', 'hwy']].isna().sum()
````

### Q2. df.dropna()를 이용해 hwy 변수의 결측치를 제거하고, 어떤 구동 방식의 hwy 평균이 높은지 확인

````python
# hwy 결측치 제거, drv 별 분리, hwy 평균 구하기
mpg_new.dropna(subset = ['hwy'])\
    .groupby('drv')\
    .agg(mean_hwy = ('hwy', 'mean'))
````
